### Code for Web Scrapping

```python
title = []
rating = []
duration = []
votes = []
year = []
#gross = []
for i in range(1,11):
    url = 'https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating%27'
    page = requests.get(url)
    htmlCode = page.text
    
    soup = BeautifulSoup(htmlCode)
    
    for x in soup.find_all('div', attrs={'class' : 'lister-item mode-advanced'}):

                   
        # title    
        titles = x.find('h3', attrs= {'class': 'lister-item-header'})
        if titles is None:
            title.append(np.NaN)
        else:
            title.append(titles.text)

        # ratings
        ratings = x.find('div', attrs={'class' : 'inline-block ratings-imdb-rating'})
        if ratings is None:
            rating.append(np.NaN)
        else:
            rating.append(ratings.text)

            
        # duration
        durations = x.find('span', attrs={'class' : 'runtime'})
        if durations is None:
            duration.append(np.NaN)
        else:
            duration.append(durations.text)

        
        
        #votes
        vote = x.find('span', attrs={'name' : 'nv'})
        if vote is None:
            votes.append(np.NaN)
        else:
            votes.append(vote.text)
            
        #year
        years = x.find('span', attrs={'class' : 'lister-item-year text-muted unbold'})
        if years is None:
            year.append(np.NaN)
        else:
            year.append(years.text)
            
        #gross
#         gros = x.find('p', attrs={'class' : 'sort-num_votes-visible'}
#         if gros is None:
#             gross.append(np.NaN)
#         else:
#             gross.append(years.text)
```

```python
print(len(title))
print(len(rating))
print(len(duration))
print(len(votes))
print(len(year))
```

### Extracting the Details using Regex

```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv("imdb1000.csv")

df.head()
```

```python
regex = r'[^\n\d,().]+'

df['Title'] = df['Title'].apply(lambda x : re.findall(regex, x))
```

```python
regex =  r'[0123456789]+'

df['Year'] = df['Year'].apply(lambda x : re.findall(regex, x))
```

```python
regex =  r'[0123456789]+'

df['Votes'] = df['Votes'].apply(lambda x : re.findall(regex, x))
```
